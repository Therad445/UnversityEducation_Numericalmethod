```python
import math
import matplotlib.pyplot as plt
import numpy as np

def get_diff(f):
    """Returns a function that calculates the numerical derivative of f(x) using the central difference formula."""
    return lambda x, h: (f(x + h) - f(x - h)) / (2 * h)

# Define the function f(x)
f = lambda x: math.cosh(x / 4)

# Define the value of epsilon (ξ)
eps = 1.1

# Calculate the constant M_2 for error estimation
M_2 = 1 / 16 * math.cosh(2 * eps / 4)

# Calculate the exact derivative of f(x) at ε
df = 1 / 4 * math.sinh(eps / 4)

# Define functions for calculating f_h(x) and f_2h(x)
f_h = lambda x, h: (f(x + h) - f(x)) / h
f_2h = get_diff(f)

# Task 1 ---------------------------------------------------------

print('Task 1')

delta = 10 ** -3

# Calculate the optimal step size for the given delta and M_2
h_opt = delta * 2 / M_2

print(f"eps = 10^-3")
print(f"f'(eps) ~ {f_2h(eps, h_opt)}")
print(f"f'(eps)  = {df}\n")

delta = 10 ** -6

# Recalculate the optimal step size for the new delta
h_opt = delta * 2 / M_2

print(f"eps = 10^-6")
print(f"f'(eps) ~ {f_2h(eps, h_opt)}")
print(f"f'(eps)  = {df}\n")

# Task 2 ---------------------------------------------------------

h = [1 / 2, 1 / 4, 1 / 8]
print('Task 2')
print(f"f_h  = {df - f_h(eps, h[0])}, {df - f_h(eps, h[1])}, {df - f_h(eps, h[2])}")
print(f"f_2h = {df - f_2h(eps, h[0])}, {df - f_2h(eps, h[1])}, {df - f_2h(eps, h[2])}\n")

# Task 3 ---------------------------------------------------------

print('Task 3')

x = np.array([i for i in range(30)])
e = np.array([abs(df - f_h(eps, 1/2**i)) for i in x])
delta = [M_2 * 1/2**i / 2 for i in range(1, 31)]

print(delta)
print(e)

plt.plot(x, e)
plt.grid()
plt.show()
```

```python
def f(x):
    return x**3

def trapezoidal_rule(f, a, b, n):
    h = (b - a) / n
    sum = (f(a) + f(b)) / 2
    for i in range(1, n):
        x = a + i * h
        sum += f(x)
    integral = h * sum
    return integral

def simpsons_rule(f, a, b, n):
    h = (b - a) / n
    sum = f(a) + f(b)
    for i in range(1, n):
        x = a + i * h
        if i % 2 == 0:
            sum += 2 * f(x)
        else:
            sum += 4 * f(x)
    integral = h / 3 * sum
    return integral

a = 0
b = 1
n = 100

trap_integral = trapezoidal_rule(f, a, b, n)
simp_integral = simpsons_rule(f, a, b, n)

print("Значение интеграла по формуле трапеций:", trap_integral)
print("Значение интеграла по формуле Симпсона:", simp_integral)

```

```python
import math

desired_accuracy = 10 ** -6  # Требуемая точность
dx = 0.1  # Начальное значение шага

while True:
    bessel_value = math.j0(dx)  # Вычисление значения функции Бесселя
    arctan_value = math.atan(1)  # Вычисление значения арктангенса единицы
    absolute_difference = abs(bessel_value - arctan_value)  # Вычисление абсолютной разницы

    if absolute_difference < desired_accuracy:
        break

    dx *= 0.1  # Уменьшение шага на порядок

pi_value = 1 / dx  # Вычисление значения числа π

print("Значение числа π:", pi_value)
```

```python
import numpy as np
import math


def simp(f, x):
    return (x[-1] - x[0]) / (3 * (len(x) - 1)) * (f(x[0]) + f(x[-1])
                                                  + 2 * np.sum(f(x[2:-1:2]))
                                                  + 4 * np.sum(f(x[1:-1:2])))


def trap(f, x):
    return (x[1] - x[0]) * ((f(x[0]) + f(x[-1])) / 2
                            + np.sum(f(x[1:-1])))


def show_res(res, s, t):
    s, t = round(s, 4), round(t, 4)
    diff = len(str(s)) - len(str(t))
    print(f'Simpson method: \t{str(s) + " " * max(-diff, 0)}, \t\tdelta = {res - s}')
    print(f'Trapezoidal method: \t{str(t) + " " * max(diff, 0)}, \t\tdelta = {res - t}')


def task1():
    a, b, n = 0, 1, 4
    x = np.linspace(a, b, n + 1)

    print('\nf(x) = x^3')
    f = lambda x: x ** 3
    show_res(1 / 4, simp(f, x), trap(f, x))

    print('\nf(x) = x^2')
    f = lambda x: x ** 2
    show_res(1 / 4, simp(f, x), trap(f, x))

    print('\nf(x) = x')
    f = lambda x: x / 2
    show_res(1 / 4, simp(f, x), trap(f, x))


def task2():
    # R = h^2 * M2 * (b - a) / 12
    # R < 10^-6
    # h < (10^-6 * 12/(M2 * (b - a))) ^ (1/2) = (10^-6 * 6) ^ (1/2)

    # f'' = 4*8 * x^2 * (x^2 + 1)^−3 − 2 * (x^2 + 1)^−2 < 8 * x^2 − 2
    # M2 < 24

    a, b, M2, eps = 0, 1, 24, 10 ** -6
    f = lambda x: 4 / (1 + x ** 2)

    h = np.sqrt(eps * 6)
    x = np.linspace(a, b, int((b - a) / h + 1) + 1)

    t = trap(f, x)
    print(f'Trapezoidal method: \t{t}, \t\tdelta = {abs(math.pi - t)}')


def task3():
    a, b, eps = 0, 1, 10 ** -6
    f = lambda x: 4 / (1 + x ** 2)
    R = lambda Ih, Ih2: abs(Ih - Ih2) / 3

    h = (a - b) / 2
    x1 = np.linspace(a, b, int((a - b) / h))
    x2 = np.linspace(a, b, int((a - b) / (h / 2)))

    while R(trap(f, x1), trap(f, x2)) >= eps:
        h /= 2
        x1 = np.linspace(a, b, int((a - b) / h))
        x2 = np.linspace(a, b, int((a - b) / (h / 2)))

    x = np.linspace(a, b, int((a - b) / (h / 2)))
    t = trap(f, x) * 4
    print(f'Trapezoidal method: \t{t}, \t\tdelta = {abs(math.pi - t)}')


def main():
    print('====== Task 1 ======')
    task1()

    print('\n====== Task 2 ======')
    task2()

    print('\n====== Task 3 ======')
    task3()


if __name__ == "__main__":
    main()
```
