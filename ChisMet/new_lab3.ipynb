```python
import numpy as np

def f(t):
    return np.cos(np.pi * t / 4)

t = np.array([0, 1/6, 1/3, 1/2, 2/3, 5/6])
y = f(t)

# Linear interpolation
t_interp = np.linspace(0, 5/6, 100)
y_interp = np.interp(t_interp, t, y)

# Plotting
import matplotlib.pyplot as plt

plt.plot(t_interp, f(t_interp), label='f(t)')
plt.plot(t_interp, y_interp, label='Interpolation')
plt.plot(t, y, 'o', label='Data points')
plt.legend()
plt.show()
```

```python
from scipy.interpolate import lagrange

# Lagrange interpolation
poly = lagrange(t, y)
y_poly = poly(t_interp)

# Plotting
plt.plot(t_interp, f(t_interp), label='f(t)')
plt.plot(t_interp, y_poly, label='Lagrange interpolation')
plt.plot(t, y, 'o', label='Data points')
plt.legend()
plt.show()
```
